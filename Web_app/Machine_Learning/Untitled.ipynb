```python
import pandas as pd

# Load the Excel file
file_path = "Online_Survey_Responses_2.xlsx"  # Replace with your file path
df = pd.read_excel(file_path)

# Display the first few rows of the data
df
```

```python
df.info()
```

```python
df.drop(['Timestamp', 'Do you consent to participate in this survey'], axis=1, inplace=True)
```

```python
df
```

```python
# Basic exploratory data analysis (EDA) - plotting categorical distributions
import seaborn as sns
import matplotlib.pyplot as plt
df_eda=df
# Plotting the age group distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df_eda, x='1. What is your age group?')
plt.title('Age Group Distribution')
plt.show()

# Plotting satisfaction with current recommendation systems
plt.figure(figsize=(8, 6))
sns.countplot(data=df_eda, x='6. How satisfied are you with the current recommendation systems used by online retailers?')
plt.title('Satisfaction with Current Recommendation Systems')
plt.show()

# Plotting the importance of accuracy of product recommendations
plt.figure(figsize=(8, 6))
sns.countplot(data=df_eda, x='5. How important is the accuracy of product recommendations in your purchasing decisions?')
plt.title('Importance of Product Recommendation Accuracy')
plt.show()
```

```python
# Encode categorical variables
df_encoded = df.apply(lambda x: pd.factorize(x)[0]+1)

# Correlation matrix
corr_matrix = df_encoded.corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from scipy import stats
# Perform ANOVA for one variable, e.g., 'Accuracy_Importance' based on 'Age_Group'
f_value, p_value = stats.f_oneway(
    df[df['15. In your opinion, how effective are current recommendation systems at understanding and responding to your sentiments?'] == 1]['Accuracy_Importance'],
    df[df['15. In your opinion, how effective are current recommendation systems at understanding and responding to your sentiments?'] == 2]['Accuracy_Importance'],
    df[df['15. In your opinion, how effective are current recommendation systems at understanding and responding to your sentiments?'] == 3]['Accuracy_Importance'],
    df[df['15. In your opinion, how effective are current recommendation systems at understanding and responding to your sentiments?'] == 4]['Accuracy_Importance'],
    df[df['15. In your opinion, how effective are current recommendation systems at understanding and responding to your sentiments?'] == 5]['Accuracy_Importance']
)

print('ANOVA F-value:', f_value)
print('ANOVA p-value:', p_value)
```
